```python
import math
import matplotlib.pyplot as plt
import numpy as np



def v(t_values): 
    return np.sin(-t_values)
def m(t_values): 
    return 1/(np.cos(-t_values))
def x(t_values): 
    return np.cos(-t_values) -1

t = np.linspace(0, 5*np.pi , 500)    

Z1 = v(t)
Z2 = m(t)
Z3 = x(t)
plt.plot(t, Z1)
plt.plot(t, Z2)
plt.plot(t, Z3)
plt.ylim(-3, 3)
plt.xlim(0,13)


```
